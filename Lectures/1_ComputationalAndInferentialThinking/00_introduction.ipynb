### Following along

All of the lecture and breakout materials are on [Github](https://github.com) at [https://github.com/profjsb/ml_course](https://github.com/profjsb/ml_course). 

The preferred option is that you use [MyBinder](https://mybinder.org/). MyBinder is a hosted service (from UC Berkeley!) that turns Jupyter notebooks in GitHub repositories into executable notebooks. MyBinder manages the dependencies for you by building a [Docker](http://docker.org) image for you and running a container of that image on demand.

Clicking [![Binder](https://mybinder.org/badge.svg)](https://mybinder.org/v2/gh/profjsb/ml_course/master) will take you to a running version this course.  You'll see something like this:

<img src="https://media.giphy.com/media/lMlwtDDwvT8qFYJvSa/giphy.gif">

You can navigate to the `Lectures/1_ComputationalAndInferentialThinking/00_introduction.ipynb` notebook
